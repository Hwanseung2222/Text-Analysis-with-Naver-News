```python
import pandas as pd  
import urllib.request  
import datetime 
import time 
import json 
import urllib.request
from bs4 import BeautifulSoup 
import re 
import os
import sys
import html
```

## 뉴스기사 본문
- 광주, 대구, 대전, 부산, 서울, 울산, 인천, 제주, 대한민국, 한국

```python
base_url = 'https://openapi.naver.com/v1/search/news.json' 
querylist = ['서울','부산','대구','인천','광주', '대전', '울산', '제주','대한민국','한국'] #8개 지역과 대한민국+한국
```

```python
for query in querylist:
    encQuery = urllib.parse.quote(query)
    n_display = 100 # 100개씩.
    start = 1
    sort = 'sim'
    url = f'{base_url}?query={encQuery}&display={n_display}&start={start}&sort={sort}'
    
    client_id = "" #각각의 cliend 고유 id
    client_secret = ""       #각각의 cline 고유 secret
    
    my_request = urllib.request.Request(url)
    my_request.add_header("X-Naver-Client-Id",client_id)
    my_request.add_header("X-Naver-Client-Secret",client_secret)
    
    response = urllib.request.urlopen(my_request)
    
    rescode = response.getcode()
    if(rescode==200):
        response_body = response.read()
    else:
        print("Error Code:" + rescode)

    search_result_str = response_body.decode('utf-8')
    search_results = json.loads(search_result_str)
    
    artice_ids = ['newsct_article','articeBody']
    
    # 네이버 뉴스 link 가 있는 뉴스만 검색하기
    titles = []
    links = []
    pubdates = []
    contents = []

    p = re.compile('https://n.news.naver.com/.+')
    for i, item in enumerate(search_results['items']):
        if p.match(item['link']):
            title = item['title']          #title 가져오기
            link = item['link']            #link 가져오기
            pubdate = item['pubDate']      #putbdate 가져오기
            titles.append(title)
            links.append(link)
            pubdates.append(pubdate)

            html = urllib.request.urlopen(link)
            bs_obj = BeautifulSoup(html, 'html.parser')
            for article_id in artice_ids:
                print(article_id)
                content = bs_obj.find_all('div',{'id':article_id})
                if len(content) > 0 :
                    contents.append(content[0].text)
                    break
                else:
                    continue
            time.sleep(3)
    

    result_dict = {'title':titles , 'link':links, 'pubdate':pubdates,'content':contents} #dict 화
    dt_now = datetime.datetime.now() 
    dt_now_date = dt_now.date() # 현재날짜 
    df = pd.DataFrame.from_dict(result_dict)
    df.to_csv(f'{dt_now_date}_{query}.csv',encoding='utf-8-sig')
    # 매번 생성되는 데이트를 그 당시에 시간과 주제에 맞게 csv 추출.
```
